```python
import pandas as pd

df_mat = pd.read_csv("./student_performance/student/student-mat.csv", sep=';')

df_por = pd.read_csv("./student_performance/student/student-por.csv", sep=';')


print(len(df_mat))
print(len(df_por))

print(df_mat.isnull().sum())
print(df_por.isnull().sum())

print(df_mat.head())
print(df_por.head())

```

```python
from sklearn.model_selection import train_test_split

df_mat_encoded = pd.get_dummies(df_mat, drop_first=True)
df_por_encoded = pd.get_dummies(df_por, drop_first=True)

X_mat = df_mat_encoded.drop(columns=['G1', 'G2', 'G3'])
y_mat = df_mat_encoded['G3']

X_por = df_por_encoded.drop(columns=['G1', 'G2', 'G3'])
y_por = df_por_encoded['G3']

X_mat_train, X_mat_test, y_mat_train, y_mat_test = train_test_split(
    X_mat, y_mat,
    test_size=0.2,
    random_state=1
)

X_por_train, X_por_test, y_por_train, y_por_test = train_test_split(
    X_por, y_por,
    test_size=0.2,
    random_state=1
)

print(df_mat_encoded.dtypes)
```
